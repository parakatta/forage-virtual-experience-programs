```python
def plot_continuous_distribution(data: pd.DataFrame = None, column: str = None, height: int = 8):
  _ = sns.displot(data, x=column, kde=True, height=height, aspect=height/5).set(title=f'Distribution of {column}');

def get_unique_values(data, column):
  num_unique_values = len(data[column].unique())
  value_counts = data[column].value_counts()
  print(f"Column: {column} has {num_unique_values} unique values\n")
  print(value_counts)

def plot_categorical_distribution(data: pd.DataFrame = None, column: str = None, height: int = 8, aspect: int = 2):
  _ = sns.catplot(data=data, x=column, kind='count', height=height, aspect=aspect).set(title=f'Distribution of {column}');

def correlation_plot(data: pd.DataFrame = None):
  corr = data.corr()
  corr.style.background_gradient(cmap='coolwarm')
```

```python
data1 = pd.read_csv('/content/sales.csv')
data2 = pd.read_csv('/content/sensor_stock_levels.csv')
data3 = pd.read_csv('/content/sensor_storage_temperature.csv')
data1.head()
```

```python
data1.drop(columns=["Unnamed: 0"], inplace=True, errors='ignore')
data2.drop(columns=["Unnamed: 0"], inplace=True, errors='ignore')
data3.drop(columns=["Unnamed: 0"], inplace=True, errors='ignore')
```

```python
data2.info()
```

```python
data3.info()
```

```python
def convert_to_datetime(data, column):
  dummy = data.copy()
  dummy[column] = pd.to_datetime(dummy[column], format = '%Y-%m-%d %H:%M:%S')
  return dummy
```

```python
data1 = convert_to_datetime(data1, 'timestamp')
data2 = convert_to_datetime(data2, 'timestamp')
data3 = convert_to_datetime(data3, 'timestamp')
data2.info()
```

```python
from datetime import datetime

def convert_timestamp_to_hourly(data: pd.DataFrame = None, column: str = None):
  dummy = data.copy()
  new_ts = dummy[column].tolist()
  new_ts = [i.strftime('%Y-%m-%d %H:00:00') for i in new_ts]
  new_ts = [datetime.strptime(i, '%Y-%m-%d %H:00:00') for i in new_ts]
  dummy[column] = new_ts
  return dummy
```

```python
data1 = convert_timestamp_to_hourly(data1, 'timestamp')
data2 = convert_timestamp_to_hourly(data2, 'timestamp')
data3 = convert_timestamp_to_hourly(data3, 'timestamp')
data2.info()
```

```python
sales_agg = data1.groupby(['timestamp', 'product_id']).agg({'quantity': 'sum'}).reset_index()
sales_agg.head()
```

```python
stock_agg = data2.groupby(['timestamp', 'product_id']).agg({'estimated_stock_pct': 'mean'}).reset_index()
stock_agg.head()
```

```python
temp_agg = data3.groupby(['timestamp']).agg({'temperature': 'mean'}).reset_index()
temp_agg.head()
```

```python
merge_df = sales_agg.merge(stock_agg, on = ['timestamp', 'product_id'], how = 'left')
merge_df.head()
```

```python
merge_df = merge_df.merge(temp_agg, on = 'timestamp', how = 'left')
merge_df.head()
```

```python
sales_other = data1[['product_id', 'category']]
sales_other = sales_other.drop_duplicates()
product_other = data1[['product_id', 'unit_price']]
product_other = product_other.drop_duplicates()
```

```python
merge_df = merge_df.merge( sales_other,  on="product_id", how="left")
merge_df = merge_df.merge( product_other,  on="product_id", how="left")
merge_df.head()
```

```python
merge_df['estimated_stock_pct'] = merge_df['estimated_stock_pct'].fillna(0)
merge_df['quantity'] = merge_df['quantity'].fillna(0)
```

```python
merge_df.head()
```

```python
merge_df.info()
```

```python
merge_df['timestamp_day_of_month'] = merge_df['timestamp'].dt.day
merge_df['timestamp_day_of_week'] = merge_df['timestamp'].dt.dayofweek
merge_df['timestamp_hour'] = merge_df['timestamp'].dt.hour
merge_df.drop(columns=['timestamp'], inplace=True)
merge_df.head()
```

```python
merge_df.drop(columns=['product_id'], inplace=True)
merge_df = pd.get_dummies(merge_df, columns=['category'])
merge_df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
```

```python
X = merge_df.drop(['estimated_stock_pct'], axis = 1)
y = merge_df['estimated_stock_pct']
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y , test_size = 0.3, random_state = 0)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
regressor = RandomForestRegressor()
regressor.fit(x_train, y_train)
y_pred = regressor.predict(x_test)
```

```python
df = pd.DataFrame({'y_pred': y_pred, "y_test": y_test})
```

```python
df
```

```python
features = [i.split("__")[0] for i in merge_df.columns]
importances = regressor.feature_importances_
indices = np.argsort(importances)

fig, ax = plt.subplots(figsize=(10, 20))
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```
